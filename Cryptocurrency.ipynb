```python
# Import Dependencies
import pandas as pd
import os
```

```python
bitcoin_csv = os.path.join("Resources", "bitcoin_cash_price.csv")
dash_csv = os.path.join ("Resources", "dash_price.csv")
```

```python
bitcoin_df = pd.read_csv(bitcoin_csv).set_index("Date")
dash_df = pd.read_csv(dash_csv).set_index("Date")
```

```python
bitcoin_df.head()
```

```python
dash_df.head()
```

```python
df = pd.merge(bitcoin_df,dash_df, left_index=True, right_index=True, suffixes = ("_bitcoin", "_dash"))
```

```python
df.head()
```

```python
Market = df[["Open_bitcoin", "Open_dash", "Close_bitcoin", "Close_dash"]].max()
```

```python
Total_value = df[["Volume_bitcoin", "Volume_dash"]].sum()
```

# Cryptocurrency summary

```python
frames = [Market, Total_value]

df = pd.concat(frames)
```
